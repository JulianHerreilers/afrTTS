```python
#modified from https://www.youtube.com/watch?v=PPmNYwVbcts#
n_fft=2048

f_s = 16000
filterbanks = librosa.filters.mel(sr=f_s, n_fft=n_fft, n_mels=80,fmin=0, fmax=f_s)
filterbanks /= -np.log(np.max(filterbanks,axis=-1)[:, None])
f = np.arange(filterbanks.T.shape[0])/n_fft*f_s
plt.figure(figsize=(9,5))
plt.xlim(right=8000)
plt.plot(f, filterbanks.T)
plt.xlabel("frequency (Hz)")
plt.ylabel("logarithmic filter magnitude")
plt.savefig('../mel_filter_explanation.pdf')
```

```python
import librosa.display
#
filename="oldies/afr_2418_5053850283.wav"
y, fs = librosa.load(filename,sr=16000)
mel_spec = librosa.feature.melspectrogram(y/1.0, sr = fs, n_mels=80,hop_length = 200)
D = librosa.stft(y, hop_length=200)  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
fig, ax = plt.subplots()
plt.xlim(left= 0)
plt.xlim(right= 3.5)
img = librosa.display.specshow(S_db, x_axis='time', y_axis='log', ax=ax,cmap='viridis')
plt.figure()



fig.colorbar(img, ax=ax, format="%+2.f dB")
plt.colorbar()
```
